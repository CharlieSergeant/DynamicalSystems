Bead on Hoop, energy terms

Hoop is spinning around $\omega$

bead swinging up or down $\dot{\theta}$

$$
r=R
$$

$$
K_{\theta}= \frac{1}{2} m r \dot{\theta} ^2 
$$

$$
K_{\phi} = \frac{1}{2} m r^2 sin^2(\theta) \omega^2
$$



$$
V = m g y = - m g r cos(\theta)
$$

$$
L = T-V
$$

$$
L = \frac{1}{2} mr^2 sin^2(\theta) \omega^2 + \frac{1}{2} m r^2 \dot{\theta}^2 + m g r cos(\theta)
$$

$$
\frac{\partial L}{\partial \dot{\theta}} = m r^2 \dot{\theta}
$$

$$
\frac{\partial L}{\partial {\theta}}
=  mr^2 sin(\theta) cos(\theta) \omega^2 
- m g r sin(\theta)
$$

$$
\frac{d}{dt} \frac{\partial L}{\partial \dot{\theta}} = m r^2 \ddot{\theta}
$$

$$
\frac{d}{dt} \frac{\partial L}{\partial \dot{q}}
-
\frac{\partial L}{\partial q}
=
0
$$

$$
m r^2 \ddot{\theta} = mr^2 sin(\theta) cos(\theta) \omega^2 
- m g r sin(\theta)
$$

$$
  \ddot{\theta} =  sin(\theta) cos(\theta) \omega^2 
-  \frac{g}{r} sin(\theta)
$$

$$
 \frac{d}{dt} \dot{\theta} =  sin(\theta) cos(\theta) \omega^2 
-  \frac{g}{r} sin(\theta)
$$

$$
\frac{d}{dt} \theta = \dot{\theta}
$$


1. Given T and V and constants
2. partial w respect to $\theta$ and $\dot{\theta}$ where q is theta

$$
\frac{\partial L}{\partial \dot{q}}
-
\frac{\partial L}{\partial q}
=
0
$$

3. partial w respect to t on $\dot{\theta}$ and set up system of 2 1st order ODEs

$$
 \frac{d}{dt} \dot{\theta} =  sin(\theta) cos(\theta) \omega^2 
-  \frac{g}{r} sin(\theta)
$$

$$
\frac{d}{dt} \theta = \dot{\theta}
$$


```
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt


r = 1
omega = 10
c = 5
g = 9.8

def f(yvec,t):
  theta = yvec[0]
  theta_dot = yvec[1]
  return np.array([theta_dot,np.sin(theta)*np.cos(theta)*omega*omega - (g/r)*np.sin(theta) - c*theta_dot])

theta_0 = 0.01
theta_dot_0 = 0
y0 = [theta_0,theta_dot_0]
tspan = [0,10]
t_eval = np.linspace(tspan[0],tspan[1],1000)

sol_odeint = odeint(f,y0,t_eval,rtol=1.0e-10,atol=1.0e-10)
```

```
plt.plot(t_eval,sol_odeint[:,0])
```
