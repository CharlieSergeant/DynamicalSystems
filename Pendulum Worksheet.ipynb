```
import numpy as np

theta0 = 3.0
thetadot0 = 0.0
g = 9.8
l = 1

y0 = np.array([thetadot0, theta0])

def roc(y):
  thetadot = y[0]
  theta = y[1]
  return(np.array([-g/l*np.sin(theta), thetadot]))

theta_plot = []
t_plot = []

y = y0
t = 0
tfinal = 10
h = 0.0001
while(t<tfinal):
  theta_plot.append(y[1])
  t_plot.append(t)
  y = y + roc(y)*h
  t = t + h

import matplotlib.pyplot as plt
plt.plot(t_plot,theta_plot)
plt.show()
```
