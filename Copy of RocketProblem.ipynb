```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```
g = 9.8
m = 1.5 # mass of rocket
lam = 0.07
ydot0 = 50 #initial speed
y0 = 0 #initial height
```

$$
\ddot{y} = f(y,\dot{y})
$$

$$
\vec{y} = \begin{bmatrix}
y \\
\dot{y}
\end{bmatrix}
$$

$$
\frac{d}{dt} \vec{y} =\begin{bmatrix}
\dot{y} \\
f(y,\dot{y})
\end{bmatrix}
$$

```
def dynamics(yvec,t):
  y = yvec[0]
  v = yvec[1]
  if(y<0):
    return np.array([0,0]) ## hard coded ground
  return np.array(
      [v,
       -g-(lam)/(m+lam*y)*v**2]
  )
```

```
yv0 = np.array([y0,ydot0])
tspan = np.linspace(0,20,10000)
sol = odeint(dynamics,yv0,tspan)
```

```
plt.plot(tspan,sol[:,0])
plt.show()
```
