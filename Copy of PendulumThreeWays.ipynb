$$
\ddot{\theta} = - \frac{g}{l} sin(\theta)
$$

```
## small angle approximation
```

$$
\ddot{\theta} \approx - \frac{g}{l} \theta
$$

$$
\theta(t) = A sin(\omega t + \phi)
$$

$$
\omega = \sqrt{\frac{g}{l}}
$$

```
## Eulers method
```

$$
NEW = OLD + CHANGE
$$

$$
CHANGE = R.O.C. \times \Delta t
$$

$$
\frac{d}{dt} \dot{\theta} = - \frac{g}{l} \theta
$$

$$
\frac{d}{dt} \theta = \dot{\theta}
$$

$$
\frac{d}{dt} [\dot{\theta}, \theta] = [-\frac{g}{l}\theta, \dot{\theta}]
$$

```
import numpy as np

theta0 = .1
thetadot0 = 0.0
g = 9.8
l = 1

y0 = np.array([thetadot0, theta0])

def roc(y):
  thetadot = y[0]
  theta = y[1]
  return(np.array([-g/l*theta, thetadot]))

theta_plot = []
t_plot = []

y = y0
t = 0
tfinal = 10
h = 0.0001
while(t<tfinal):
  theta_plot.append(y[1])
  t_plot.append(t)
  y = y + roc(y)*h
  t = t + h

import matplotlib.pyplot as plt
plt.plot(t_plot,theta_plot)
plt.show()
```
